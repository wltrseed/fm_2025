```python
import math

def f(a,b,c):
    x = a * math.sin(math.pi * b / 180) / c
    y = a ** 2 * math.sin(math.pi * b / 180) ** 2 / 2 / c
    z = a ** 2 * math.sin(2 * math.pi * b / 180) / c
    return y * z / 2
```

```python
# как это должно выглядеть чтобы было понятно

def triangle_area(v_0: float, alpha: float, g: float) -> float:
    """
    Функция вычисляет примерную площадь под траекторией 
        брошенного под углом к горизонту объекта

    :param v_0: начальная скорость брошенного объекта
    :type v_0: float
    :param alpha: угол броска в градусах
    :type alpha: float
    :param g: ускорение свободного падения
    :type g: float
    :return: площадь треугольника с основанием длины проекции траектории на землю, умноженной на высоту
    :rtype: float
    """
    assert v_0 > 0, 'Скорость должна быть больше 0!'
    assert alpha > 0, 'Угол броска должен быть больше 0!'
    assert g > 0, 'Ускорение свободного падения должно быть больше 0!'
    
    # привести градусы к радианам
    radian_angle = math.pi * alpha / 180
    
    # рассчитать время полета, максимальный подъем, расстояние до места падения
    hovering_time: float = v_0 * math.sin(radian_angle) / g
    max_height: float = v_0 ** 2 * math.sin(radian_angle) ** 2 / 2 / g
    distance: float = v_0 ** 2 * math.sin(2 * radian_angle) / g
    
    # рассчитать аппроксимацию площади фигуры описанной дугой траектории как площадь треугольника
    # с высотой броска и основанием как расстоянием до падения
    triangle_area: float = max_height * distance / 2
    return triangle_area
```

Функция должна:
- Принимать коэффициенты a, b, c как float
- Возвращать кортеж с корнями уравнения

- Использовать комплексные числа для отрицательного дискриминанта
- Включать проверки входных данных


```python
import math
from typing import Tuple, Union
import cmath

def calculate_quadratic_roots(a: float, b: float, c: float) -> Union[Tuple[float, float], Tuple[complex, complex], None]:
    if not all(isinstance(coef, (int, float)) for coef in [a, b, c]):
        return None
    if a == 0:
        return None  
    discriminant = b**2 - 4 * a * c    
    if discriminant > 0:
        sqrt_d = math.sqrt(discriminant)
        x1 = (-b + sqrt_d) / (2 * a)
        x2 = (-b - sqrt_d) / (2 * a)
        return (x1, x2)  
    elif discriminant == 0:
        x = -b / (2 * a)
        return (x, x)   
    else:
        sqrt_d = cmath.sqrt(discriminant)
        x1 = (-b + sqrt_d) / (2 * a)
        x2 = (-b - sqrt_d) / (2 * a)
        return (x1, x2)
# Тесты для проверки
print("Тесты для calculate_quadratic_roots:")
print(f"x² - 5x + 6 = 0: {calculate_quadratic_roots(1, -5, 6)}")  # Ожидается (3.0, 2.0)
print(f"x² - 4x + 4 = 0: {calculate_quadratic_roots(1, -4, 4)}")  # Ожидается (2.0, 2.0)
print(f"x² + x + 1 = 0: {calculate_quadratic_roots(1, 1, 1)}")    # Ожидается комплексные корни
print(f"0x² + 2x + 1 = 0: {calculate_quadratic_roots(0, 2, 1)}")  # Ожидается None (не квадратное)
```

### Задание 4: Работа со строками

Создайте функцию `format_person_info`, которая форматирует информацию о человеке в красивую строку.

Функция должна:
- Принимать имя, возраст, профессию и зарплату
- Форматировать строку с выравниванием и ограничением точности для зарплаты
- Обрабатывать случаи когда данные могут быть None
- Возвращать отформатированную строку


```python
from typing import Optional

def format_person_info(name: Optional[str], age: Optional[int], 
                      profession: Optional[str], salary: Optional[float]) -> str:
    
    name_str = name if name is not None else "Не указано"
    age_str = str(age) if age is not None else "Не указан"
    profession_str = profession if profession is not None else "Не указана"

    info_lines = []
    
    if salary is not None:
        salary_str = f"{salary:,.2f} руб.".replace(',', ' ')
    else:
        salary_str = "Не указана"
    
    info_lines = [
    "=" * 40,
    "Информация о человеке",
    "=" * 40,
    f"Имя:        {name_str}",
    f"Возраст:    {age_str}",
    f"Профессия:  {profession_str}",
    f"Зарплата:   {salary_str}",
    "=" * 40
    ]
    return "\n".join(info_lines)

# Тесты для проверки
print("Тесты для format_person_info:")
print(format_person_info("Иван Петров", 30, "Программист", 150000.567))
print("\n" + "="*50 + "\n")
print(format_person_info("Мария", 25, None, 120000))
print("\n" + "="*50 + "\n")
print(format_person_info(None, 35, "Дизайнер", None))
print("\n" + "="*50 + "\n")
print(format_person_info("Алексей", None, "Менеджер", 95000.123))
```

### Задание 5: Комплексная задача

Создайте функцию `process_data`, которая обрабатывает список строк с данными о студентах и возвращает статистику.

Каждая строка имеет формат: "Имя,Возраст,Оценка1,Оценка2,Оценка3"

Функция должна:
- Парсить строки и извлекать данные
- Вычислять среднюю оценку для каждого студента
- Находить лучшего и худшего студента по средней оценке
- Возвращать результат в виде словаря с ключами: 'best_student', 'worst_student', 'average_grade'
- Обрабатывать ошибки парсинга (пропускать некорректные строки)


```python
from typing import List, Dict, Any

# Ваше решение здесь
def process_data(student_data: List[str]) -> Dict[str, Any]:
    """
    Обрабатывает данные о студентах и возвращает статистику.
    
    :param student_data: список строк с данными в формате "Имя,Возраст,Оценка1,Оценка2,Оценка3"
    :return: словарь со статистикой
    """
    students = []
    
    for line in student_data:
        if not line.strip():
            continue
            
        parts = line.split(',')
        
        if len(parts) != 5:
            continue
            
        try:
            name = parts[0].strip()
            age = int(parts[1].strip())
            grades = [int(part.strip()) for part in parts[2:5]]
            
            # Проверяем, что оценки в допустимом диапазоне (2-5)
            if not all(2 <= grade <= 5 for grade in grades):
                continue
                
            # Вычисляем среднюю оценку для студента
            average_grade = sum(grades) / len(grades)
            
            students.append({
                'name': name,
                'age': age,
                'grades': grades,
                'average_grade': average_grade
            })
            
        except (ValueError, IndexError):
            continue
    
    if not students:
        return {
            'best_student': None,
            'worst_student': None,
            'average_grade': 0
        }
    
    best_student = max(students, key=lambda x: x['average_grade'])
    worst_student = min(students, key=lambda x: x['average_grade'])
    
    total_average = sum(student['average_grade'] for student in students) / len(students)
    
    return {
        'best_student': best_student['name'],
        'worst_student': worst_student['name'],
        'average_grade': round(total_average, 2)
    }

# Тестовые данные
test_students = [
    "Иван Петров,20,5,4,5",
    "Мария Сидорова,19,3,4,3",
    "Алексей Козлов,21,5,5,5",
    "Анна Морозова,20,4,3,4",
    "Некорректная строка",
    "Петр Иванов,22,2,3,2"
]

# Тесты для проверки
print("Тесты для process_data:")
result = process_data(test_students)
print(f"Результат: {result}")

# Ожидаемый результат должен содержать:
# - best_student: студент с наивысшей средней оценкой
# - worst_student: студент с наименьшей средней оценкой  
# - average_grade: общая средняя оценка всех студентов
```
